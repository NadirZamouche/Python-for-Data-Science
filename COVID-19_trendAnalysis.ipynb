```python
df2 = df.groupby(['Country','Date'])[['Confirmed','Deaths','Recovered']].sum().reset_index()
df2
```

```python
df3 = df2[df2['Confirmed']>1].reset_index()
df3
```

```python
countries = df2['Country'].unique()
len(countries)
```

```python
import matplotlib.pyplot as plt
for idx in range(0,len(countries)):    
    C = df2[df2['Country']==countries[idx]].reset_index()        
    plt.scatter(np.arange(0,len(C)),C['Confirmed'],color='orange',label='Confirmed')
    plt.scatter(np.arange(0,len(C)),C['Recovered'],color='green',label='Recovered')
    plt.scatter(np.arange(0,len(C)),C['Deaths'],color='red',label='Deaths')
    plt.title(countries[idx])
    plt.xlabel('Days since the first suspect')
    plt.ylabel('Number of cases')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
bar_width = 1
for idx in range(0,len(countries)):    
    C = df2[df2['Country']==countries[idx]].reset_index()    
    x = np.arange(0,len(C))
    plt.bar(x,C['Confirmed'],color='orange',label='Confirmed')
    plt.bar(x + bar_width,C['Recovered'],color='green',label='Recovered')
    plt.bar(x + 2*bar_width,C['Deaths'],color='red',label='Deaths')
    plt.title(countries[idx])
    plt.xlabel('Days since the first suspect')
    plt.ylabel('Number of cases')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
bar_width = 0.2  # Width of each bar
index = np.arange(len(countries))  # Index for the x-axis
fig, ax = plt.subplots(figsize=(14, 10))  # Adjust the figure size as needed

for idx in range(0, len(countries)):    
    C = df2[df2['Country'] == countries[idx]].reset_index()        
    # Add the values to the corresponding bars
    ax.bar(index[idx], C['Confirmed'], bar_width, color='orange', label='Confirmed')
    ax.bar(index[idx] + bar_width, C['Recovered'], bar_width, color='green', label='Recovered')
    ax.bar(index[idx] + (2 * bar_width), C['Deaths'], bar_width, color='red', label='Deaths')
ax.set_xlabel('Countries')
ax.set_ylabel('Number of cases')
ax.set_title('COVID-19 Statistics by Country')
# Rotating x-axis labels for better visibility
ax.set_xticks(index + (bar_width))
ax.set_xticklabels(countries, rotation=5, ha='right')
ax.legend(['Confirmed', 'Recovered', 'Deaths'])
plt.tight_layout()  # Ensures all elements fit within the figure
plt.grid()
plt.show()
```

```python
for country in countries:
    C = df2[df2['Country'] == country].reset_index()
    date = C['Date'].iloc[0]
    if C['Confirmed'].iloc[0] == 0 or C['Recovered'].iloc[0] == 0 or C['Deaths'].iloc[0] == 0:
        print(f"No meaningful data available for {country} on {date}. Skipping pie chart.")
    else:
        labels = ['Confirmed', 'Recovered', 'Deaths']
        sizes = [C['Confirmed'].iloc[0], C['Recovered'].iloc[0], C['Deaths'].iloc[0]]
        colors = ['orange', 'green', 'red']
        fig, ax = plt.subplots(figsize=(8, 8))
        ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
        ax.set_title(f'COVID-19 Statistics for {country} on {date}')
        plt.tight_layout()
        plt.legend()
        plt.show()
```

```python
# Initialize variables to store data for all countries
labels = []
sizes = []
MyColors = np.array(['red','blue','green','yellow','orange','black','cyan','purple'])

# Loop through each country
for country in countries:
    C = df2[df2['Country'] == country].reset_index()
    date = C['Date'].iloc[0]
    confirmed_cases = C['Confirmed'].iloc[0]
    labels.append(country)
    sizes.append(confirmed_cases)

# Use specific colors from MyColors array
colors = MyColors[:len(labels)]

# Create the pie chart
fig, ax = plt.subplots(figsize=(14, 14))
plt.pie(sizes, labels=[''] * len(labels), colors=colors, autopct='', startangle=140)
ax.set_title(f'COVID-19 Confirmed Cases by Country')
plt.tight_layout()
plt.legend(labels=[f'{label} ({size}%)' for label, size in zip(labels, sizes)], loc='best')
plt.grid()
plt.show()
```
