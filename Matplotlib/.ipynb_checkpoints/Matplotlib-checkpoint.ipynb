# Matplotlib
## 1. Intro to Matplotlib
Matplotlib is a Python library used for creating static, animated, and interactive visualizations in a wide variety of formats. It provides a high-level interface for drawing attractive and informative statistical graphics.
* We can import like this: "import matplotlib.pyplot as plt" better than this: "from matplotlib import pyplot as plt".
* Here are some useful funcs:
  - np.linspace(start,end,number): returns evenly spaced numbers over a specified interval from starting nmber to the end and the last one represents the number of points to be generated.
  - plt.plot(x,y): to draw line graphs. Side note: for y aka f(x) we usually use numpy here.
  - plt.scatter(x,y): to draw point graphs and we can specify a range o values to be represented as we saw in numpy arrrays plt.scatter(x[range],y[samerange]).
  - We can specify the color too plt.scatter(x,y,color='colorname').
  - plt.title("Grapgh title").
  - plt.xlabel("title under x") & plt.ylabel("title under y").
  - plt.legend(): to show what corresponds to what colors, infos ...etc but I need to specify the colors and label names for each plot I draw also if we have a loop but all in one grah specify the legend this way: legend(['Confirmed', 'Recovered', 'Deaths']) & we can make specific legend as well like this: plt.legend(labels=[f'{label} ({size}%)' for label, size in zip(labels, sizes)], loc='best').
  - plt.grid(): to toggle the visibility of the grid lines on the current axes for all types of charts besides pie chart.
  - plt.hist(x): to draw a histogram which is a graph showing frequency distributions (the number of observations within each given interval).
  - plt.bar(x,y): to draw bar graphs and for each bar after in x add 1 for width.
  - plt.pie(y): to draw pie charts.
    * To show numbers on pie chart: autopct='%1.1f%%'.
    * To hide labels: labels=[''] * len(labels).
  - plt.tight_layout(): a really good practise used to automatically adjust the subplots and other elements of a figure so that they fit within the figure area without overlapping.
  - plt.show(): to execute & show the whole stuff.
#### Note 1: in fig size always keep the difference of 4: figsize(N,N-4).
#### Note 2: to use specific color from a custom list of colors: colors = MyColors[:len(labels)] or from random unique ones: colors = plt.cm.tab10(np.arange(len(labels))).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,1000)
y = np.sin(x)
plt.plot(x,y)
plt.grid()
```

```python
plt.scatter(x,y)
plt.grid()
```

```python
plt.scatter(x[:30],y[:30])
plt.grid()
```

```python
plt.scatter(x[450:500],y[450:500])
plt.grid()
```

```python
plt.scatter(x[::10],y[::10])
plt.grid()
```

```python
plt.scatter(x[::10],y[::10], color='red')
plt.grid()
```

```python
plt.plot(x,y, color='b')
plt.plot(x,np.cos(x), color='g')
plt.grid()
```
